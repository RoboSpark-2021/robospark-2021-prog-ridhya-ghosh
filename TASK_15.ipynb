```python
from sklearn.preprocessing import LabelEncoder
```

```python
l = ['embarked','home.dest','boat']
for i in l:
    le = LabelEncoder()
    le.fit(df[i])
    df[i]=le.transform(df[i])
```

```python
df
```

```python
col_null = ['embarked','fare','body','boat'] 

med = []


for i in col_null: 
    med.append(df[i].median())  

k = 0 

for i in col_null:
    df[i].fillna(med[k] , inplace = True)   
    k = k + 1 
```

```python
df
```

```python
df.isnull().sum()

```

```python
col_norm = ['fare']
Max = [] 

for i in col_norm : 
    Max.append(df[i].max())

k = 0 
 
for i in col_norm : 
    df[i] = df[i]/Max[k]
    k = k + 1 
```

```python
x = df.drop(['survived'], axis=1)
y = df[['survived']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
d = DecisionTreeClassifier()
d.fit(X_train, y_train)
```

```python
acc_score = d.score(X_test,y_test)
```

```python
print("Accuracy of my model is: ",acc_score*100)
```
