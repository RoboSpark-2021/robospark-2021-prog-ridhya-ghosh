```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('C:/Users/Asus/Downloads/dementia_data.csv')
```

```python
#1
df_copy = df.copy()
vis_copy = df.copy()
```

```python
#2
le1 = LabelEncoder()
le2 = LabelEncoder()
le1.fit(df['Group'])
le2.fit(df['M/F'])


```

```python
#2
df['Group']=le1.transform(df['Group'])
df['M/F']=le2.transform(df['M/F'])
```

```python
#3
df_copy['M/F'] = pd.get_dummies(df_copy['M/F'])
print(df_copy)
```

```python
#4 
Y = vis_copy['M/F'].value_counts(dropna=True)
```

```python
#4
X = ['Female','Male']
plt.bar(X , Y , color = ['r' , 'b'],width=0.5)
# plt.hist()
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Count of Males and Females")
plt.show()
```
